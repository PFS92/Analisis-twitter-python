```python
#Se ha creado la clase StreamListener para:

class StreamListener(tweepy.Stream):
    # Crear el fichero e inicializar el objeto que escribe en csv      
    def on_connect(self):
        self.csvfile = open('data.csv', 'w',newline = "\n",encoding='utf-8')
        self.writer = csv.writer(self.csvfile, delimiter=',', quotechar='"')
        self.writer.writerow(['Text','Date','Name','Location'])
       
        print("Conectado")
    # Recoger datos que lleguen de tweeter
    def on_status(self,status):
        print(f"Ha llegado un tweet")
        print(status.text)
        print("---")
        print(status.author.name)
        print("---")
         
        self.writer.writerow([status.text,status.created_at, status.author.name, status.author.location])
        return False
      
    # Cerrar el fichero    
    def on_disconnect(self):
        self.csvfile.close()
        
        
```

```python
#Con la liberia pandas leemos el fichero y lo convertimos a un dataframe
df = pd.read_csv('data.csv',delimiter = ',', quotechar='"',header=0)
```

```python
#Dataframe original del fichero
df
```

```python
#FunciÃ³n lambda para aÃ±adir la columna que cuente el nÂº de caracteres del texto 
df["Numero_caracter"]=df['Text'].apply(lambda x: len(x))
```

```python
df
```

```python
#FunciÃ³n lambda para aÃ±adir la columna que cuente el nÂº de palabras 
df["Numero_palabras"]=df['Text'].apply(lambda x: len(x.split(" ")))
```

```python
df
```

```python
#En este paso ya que el campo Date se trata de un str, lo convierto a tipo Date para obtener el campo hora.
df['Date']= pd.to_datetime(df.Date)
```

```python
df
```

```python
#Obtengo la hora del campo Date
df['Hora'] = df['Date'].apply(lambda x: x.hour)
```

```python
df
```

```python
#Con la funciÃ³n count obtengo el nÂº de tweet y lo agrupo por hora sacando el nÂº de tweet por hora. 
df2 = df.groupby(df['Hora']).count()
```
